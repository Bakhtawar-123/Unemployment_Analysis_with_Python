```python
import pandas as pd
```

```python
# Read the CSV file
df = pd.read_csv('Unemployment in India.csv')
```

```python
# Display the first few rows of the data
df.head()
```

```python
df.columns
```

```python
df.shape
```

# Calculate the average unemployment rate

```python
average_unemployment_rate = df[' Estimated Unemployment Rate (%)'].mean()
print('Average Unemployment Rate:', average_unemployment_rate)
```


# Calculate the total number of employed individuals

```python
total_employed = df[' Estimated Employed'].sum()
print('Total Employed:', total_employed)
```

# Calculate the average labor participation rate

```python
average_labor_participation_rate = df[' Estimated Labour Participation Rate (%)'].mean()
print('Average Labor Participation Rate:', average_labor_participation_rate)
```

# Group data by region and calculate the average unemployment rate for each region

```python
average_unemployment_by_region = df.groupby('Region')[' Estimated Unemployment Rate (%)'].mean()
print('Average Unemployment Rate by Region:\n', average_unemployment_by_region)
```

#average unemployment rate for each region graph

```python
import matplotlib.pyplot as plt
```

```python
# Create a bar plot
plt.figure(figsize=(10, 6))
average_unemployment_by_region.plot(kind='bar')
plt.title('Average Unemployment Rate by Region')
plt.xlabel('Region')
plt.ylabel('Average Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.show()
```
